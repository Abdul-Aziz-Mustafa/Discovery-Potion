```python
# Import libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for acetylcholinesterase
t = new_client.target
tq = t.search('acetylcholinesterase')
ts = pd.DataFrame.from_dict(tq)
display(ts)
```

```python
#Data for Human Acetylcholinesterase
human_target = ts.target_chembl_id[0]
display(human_target)
```

```python
#Retrieve only data for Human Acetylcholinesterase
activity = new_client.activity
frame1 = pd.DataFrame.from_dict(activity.filter(target_chembl_id=human_target).filter(standard_type="IC50"))
display(frame1)
```

```python
#Save the resulting bioactivity data to a CSV file
frame1.to_csv('Human_acetylcholinesterase_raw_data.csv', index=False)
```

```python
frame2 = frame1[frame1.standard_value.notna()]
frame2 = frame2[frame1.canonical_smiles.notna()]
frame2
```

```python
len(frame2.canonical_smiles.unique())
```

```python
frame2_nd = frame2.drop_duplicates(['canonical_smiles'])
frame2_nd
```

```python
cols = ['molecule_chembl_id','canonical_smiles','standard_value']
frame3 = frame2_nd[cols]
frame3
```

```python
frame3.to_csv('preprocessed_data.csv', index=False)
```

```python
frame4 = pd.read_csv('preprocessed_data.csv')
```

```python
bioactivity_class = []
def f(i):
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")

for i in frame4.standard_value:
  f(i)
```

```python
col = ['molecule_chembl_id','canonical_smiles','standard_value']
rf = frame4[col]
rf
```

```python
bioactivity_c = pd.Series(bioactivity_class, name='class')
frame5 = pd.concat([frame4, bioactivity_c], axis=1)
frame5
```

```python
frame5.to_csv('ACHE_bioactivity_data.csv', index=False)
```
